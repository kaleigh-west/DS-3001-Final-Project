## Read in Dataset

```python
# Import pandas
import pandas as pd
```

```python
# link to dataset info: https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition
obesity_data_raw = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
obesity_data_raw
```

```python
# Rename columns for clarity
obesity_data_renamed = obesity_data_raw.rename(columns={'FAVC': 'high_freq_high_cal_intake',
                                                           'FCVC': 'vegetable_frequency',
                                                          'NCP': 'number_meals_per_day',
                                                          'CAEC': 'food_between_meals',
                                                          'CH2O': 'water_intake',
                                                          'SCC': 'monotor_daily_calories',
                                                          'FAF': 'physical_activity_freq',
                                                          'TUE': 'tech_time',
                                                          'CALC': 'alcohol_freq',
                                                          'MTRANS': 'transportation',
                                                          'NObeyesdad': 'obesity_level'})
obesity_data_renamed
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Count NaNs
obesity_data_renamed.isna().sum()
```

```python
# See unique values in each non-numeric column

# Extract non-numeric columns
non_numeric_obesity = obesity_data_renamed.select_dtypes(exclude=['number'])

# List of all non-numeric column names to iterate through
column_names = non_numeric_obesity.columns.tolist()

# Make array of unique values in each column
unique_object_values = []
for column in column_names:
    unique_object_values.append(non_numeric_obesity[column].unique())

unique_object_values
```

```python
# Count occurrences of each obesity level grouped by gender
obesity_counts = obesity_data_renamed.groupby(['obesity_level', 'Gender']).size().unstack(fill_value=0)
obesity_counts
```

```python
# Create bar chart of obesity level counts by gender

# Define colors for each gender
color_map = {'Male': 'blue', 'Female': 'red'}

# Plot bar chart
ax = obesity_counts.plot(kind='bar', color=[color_map[col] for col in obesity_counts.columns], edgecolor='black')

# Add labels, title, and legend
ax.set_xlabel("Obesity Level")
ax.set_ylabel("Count")
ax.set_title("Obesity Level Counts by Gender")
ax.legend(title="Gender")

plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Count occurrences of each unique string
counts = obesity_data_renamed['obesity_level'].value_counts()
counts
```

```python
# Extract labels and sizes
labels = counts.index  # Unique categories
sizes = counts.values  # Corresponding counts

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title('Target Frequency Distribution')
plt.show()
```

```python
# Combine all obese/non-obese columns to make target values "Obese" and "Not Obese"

# Underweight Less than 18.5
# â€¢Normal 18.5 to 24.9
# â€¢Overweight 25.0 to 29.9
# â€¢Obesity I 30.0 to 34.9
# â€¢Obesity II 35.0 to 39.9
# â€¢Obesity III Higher than 40

# Define which levels are considered "Obese"
obese_categories = ['Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']

# Create a binary column: 1 if obese, 0 if not
obesity_data_renamed['Obese'] = obesity_data_renamed['obesity_level'].isin(obese_categories).astype(int)

obesity_data_renamed
```

```python
# Count occurrences of each new obesity level grouped by gender
obesity_counts_new = obesity_data_renamed.groupby(['Obese', 'Gender']).size().unstack(fill_value=0)
obesity_counts_new
```

```python
# Create bar chart of obesity level counts by gender

# Define colors for each gender
color_map = {'Male': 'blue', 'Female': 'red'}

# Plot bar chart
ax_new = obesity_counts_new.plot(kind='bar', color=[color_map[col] for col in obesity_counts_new.columns], edgecolor='black')

# Add labels, title, and legend
ax_new.set_xlabel("Obesity Level")
ax_new.set_ylabel("Count")
ax_new.set_title("Obesity Level Counts by Gender")
ax_new.legend(title="Gender")

plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```
