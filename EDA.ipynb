## Read in Dataset

```python
# Import pandas
import pandas as pd
```

```python
# link to dataset info: https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition
obesity_data_raw = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
obesity_data_raw
```

```python
# Rename columns for clarity
obesity_data_renamed = obesity_data_raw.rename(columns={'FAVC': 'high_freq_high_cal_intake',
                                                           'FCVC': 'vegetable_frequency',
                                                          'NCP': 'number_meals_per_day',
                                                          'CAEC': 'food_between_meals',
                                                          'CH2O': 'water_intake',
                                                          'SCC': 'monotor_daily_calories',
                                                          'FAF': 'physical_activity_freq',
                                                          'TUE': 'tech_time',
                                                          'CALC': 'alcohol_freq',
                                                          'MTRANS': 'transportation',
                                                          'NObeyesdad': 'obesity_level'})
obesity_data_renamed
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Count occurrences of each unique string
counts = obesity_data_renamed['obesity_level'].value_counts()
counts
```

```python
# Extract labels and sizes
labels = counts.index  # Unique categories
sizes = counts.values  # Corresponding counts

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title('Target Frequency Distribution')
plt.show()
```
